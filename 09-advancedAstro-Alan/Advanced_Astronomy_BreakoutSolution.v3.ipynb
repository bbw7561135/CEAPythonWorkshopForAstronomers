# Breakout Solution

```python
from astropy.io import fits
import astropy.units as u
import matplotlib.pylab as plt
from wcsaxes import WCS
import numpy as np

# Open data file
hdu_data = fits.open('CygX-3_FermiLAT.fits')
image_data = hdu_data[0].data
header_data = hdu_data[0].header
wcs = WCS(header_data) # read coordinates from FITS header

# Open model file
hdu_model = fits.open('CygX-3_Model.fits')
image_model = hdu_model[0].data
header_model = hdu_model[0].header

# Compute the residual image
residual = (image_data - image_model) / (image_model)**0.5

# Convolve with gaussian kernel
from astropy.convolution import Gaussian2DKernel, convolve
gauss = Gaussian2DKernel(stddev=2)
convolved_residual=convolve(residual, gauss, boundary='extend')

# Plot
    # Initialize the plot
fig = plt.figure()
ax = fig.add_axes([0.15, 0.1, 0.8, 0.8], projection = wcs) # use physical coordinates
lon = ax.coords['glon']
lat = ax.coords['glat']
lon.set_axislabel('Galactic longitude')
lat.set_axislabel('Galactic latitude')
lon.grid(color='black', alpha=0.5, linestyle='solid')
lat.grid(color='black', alpha=0.5, linestyle='solid')

    # Add labels and colorbar
p = ax.imshow(convolved_residual, cmap='gist_stern', origin='lower')
cbar = fig.colorbar(p)
cbar.set_label('Residual $\sigma$/pixel')
xmin, xmax = ax.get_xlim(); ymin, ymax = ax.get_ylim()

# Query Simbad
from astroquery.simbad import Simbad
from astropy.coordinates import SkyCoord
center_glon=header_data['CRVAL1'] # center of the map informations
center_glat=header_data['CRVAL2']
size=np.abs(header_data['NAXIS1'] * header_data['CDELT1']) # image size in degrees

sources_region = Simbad.query_criteria('region(box, GAL, '+str(center_glon)+' +'+
                str(center_glat)+', '+str(size*0.75)+'d '+str(size*0.75)+'d)', cat='2FGL') # takes some time...

# Plot the sources
tr_gal = ax.get_transform("galactic") # Transform coordinates into pixels'
for source in sources_region:
    gal_pos = SkyCoord(source['RA'], source['DEC'], unit=(u.hour,u.deg), frame='icrs').galactic
    ax.plot(gal_pos.l.deg, gal_pos.b.deg, color='white', marker='x', markersize=4, transform=tr_gal)
    ax.text(gal_pos.l.deg, gal_pos.b.deg, source['MAIN_ID'], color='white', size=6, transform=tr_gal)


ax.set_xlim(xmin, xmax); ax.set_ylim(ymin, ymax)

fig.savefig('breakout_plot.pdf', bbox_inches='tight')
```
